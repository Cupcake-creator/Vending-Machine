```python
class Product:
    def __init__(self, code, name, price):
        self.code = code
        self.name = name
        self.price = price
    
class VendingMachine:
    def __init__(self):
        self.amount = 0
        self.products = []
    
    def addproduct(self, item):
        self.products.append(item)
        
    def showproducts(self):
        print("\n___WELCOME TO Jew's Vending Machine___ ")
        print('\n Code     Product                           Price')
        print('_____________________________________\n')

        for product in self.products:
            print(product.code + " => " + product.name + " " + str(product.price) + "\t ฿")

        print('_____________________________________\n')
        
    def selectedproduct(self, selected):
        print('_____________________________________\n')
        for product in self.products:
            if product.code == selected:
                print(" Your choose product is " + product.name  )
                print(" ->   Total price is " + str(product.price) + " ฿")   
                return product

    def insert_cash_for_product(self, selected):
        while (self.amount < selected.price):
                print('_____________________________________\n')
                in_come = 1
                count = 0
                in_come = int(input ("Insert Banknote or Coins \n ex. 100 50 20 10 5 2 1\n remaining " + str(abs(self.amount- selected.price)) + " ฿ => "))
                
                if ((in_come == 100) or (in_come == 50) or (in_come == 20) or (in_come == 10) or (in_come == 5) or (in_come == 2) or (in_come == 1)):
                        count     = int(input ("Number of " + str(in_come) + "฿ [Banknote or Coins]  => "))
                        self.amount = self.amount + (in_come*count)
                else:
                    print('_____________________________________\n')
                    print("** Not valid input **" )
                    print('_____________________________________\n')
                    print("Insert more " + str(abs(int(self.amount - selected.price))) + "฿")

    def buy_selected(self,selected):
        print('_____________________________________\n\n')
        self.amount -= selected.price       
        print('You bought  ' + selected.name + "\n")
        print('_____________________________________\n')
        print("Your exchange money  (" + str(self.amount) + "฿)\n ")
        Cal_exchange(self.amount)
    
    def confirm(self): 
        while True:
            submit = input("Confirm Order ? (y/n) => ").lower()
            if( "y"== submit):
                return True
                break
            if( "n"== submit):
                return False
                break
            else:
                print("** Wrong input **")   
                
    def refund(self):
        print('_____________________________________\n')
        print("Your refund money  (" + str(self.amount) + "฿)\n ")
        Cal_exchange(self.amount)
        
    def end(self):
        print('\n____________THANK YOU_______________\n')
        print('____________THANK YOU_______________\n')
        print('____________THANK YOU_______________\n')
        print("/////////////////////////////////////////////////////\n\n\n\n\n")


def Cal_exchange(amount):
    while (amount > 0):
        if (amount >= 100 ):
            a = int(amount / 100)
            amount -= (a*100) 
            print (" 100 ฿ ->" + str(a))   
        elif (amount >= 50 ):
            a = int(amount / 50)
            amount -= (a*50) 
            print (" 50  ฿ ->" + str(a))
        elif (amount >= 20 ):
            a = int(amount / 20)
            amount -= (a*20) 
            print (" 20  ฿ ->" + str(a))
        elif (amount >= 10 ):
            a = int(amount / 10)
            amount -= (a*10) 
            print (" 10  ฿ ->" + str(a))
        elif (amount >= 5 ):
            a = int(amount / 5)
            amount -= (a*5) 
            print (" 5   ฿ ->" + str(a))
        elif (amount >= 2 ):
            a = int(amount / 2)
            amount -= (a*2) 
            print (" 2   ฿ ->" + str(a))
        elif (amount == 1 ):
            a = int(amount)
            amount -= 1
            print (" 1   ฿ ->" + str(a))
               
def vending():
    while True:
        Machine = VendingMachine()
        Machine.addproduct( Product(  "A1", "Coke original                   ", 12 ))
        Machine.addproduct( Product(  "A2", "Fanta grape                     ", 12 ))
        Machine.addproduct( Product(  "A3", "SHIZUOKA HOJICHA         ", 30 ))
        Machine.addproduct( Product(  "A4", "Unif High Fiber & Vitamins ", 25 ))
        Machine.addproduct( Product(  "A5", "Mineral water                   ", 12 ))
        Machine.showproducts()
        #selected = input('Insert products code: ')
        selected = Machine.selectedproduct(input('Insert products Code: '))
        while True:
            if selected:
                if(Machine.confirm()):
                    Machine.insert_cash_for_product(selected)
                    if(Machine.confirm()):
                        Machine.buy_selected(selected)
                        Machine.end()
                        break
                    else:
                        Machine.refund()
                        Machine.end()
                        break
                else:
                    break
            else:
                print("** Not found that product code **" )
                print('_____________________________________\n')
                selected = Machine.selectedproduct(input('Insert products Code: '))

vending()
```
